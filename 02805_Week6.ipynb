--- **Describing lexical diversity** ---

Lexical diversity is all about how diversely the text is written. If the text is written using a higher number of different words throughout, the text is more diverse. This is not about the quantity of a specific word occuring in the text, but how many unique words that are used.

One could use this notion to describe the different of complexity of a text or to make guesses at who might have written the text. I.e. one would expect the text of a child to have a lower level of lexical diversity since their vocabularies are not (often) as large as an adult. Then again, everything that Trump says would be thought to be a child -- to differ here one would have to look at the semantics of the text (what it actually means).

```python
# function
def percent(word,text):
    wordsInAll = 0
    numberOfWord = 0
    for words in text:
        if words.strip().lower() == word.lower():
            numberOfWord = numberOfWord + 1
        wordsInAll = wordsInAll + 1
    return "The word occurs %d number of times and make up for %f percent of the %d words in %s" % (numberOfWord, numberOfWord/wordsInAll, wordsInAll, text)
```

#### Comparing genres

```python
# The text to look for words in
text = brown.words(categories='learned')

# The frequency distribution
freq_dist2 = FreqDist([w.lower() for w in text])

# Words to look for
modals = ['okay','the','power','exciting','new']

# Printing the number of words
for m in modals:
    print m + ":", freq_dist2[m]
```

#### Conditional frequency distribution

```python
cond_fdist = nltk.ConditionalFreqDist(
            (genre, word)
            for genre in brown.categories()
            for word in brown.words(categories=genre))
# Values to work on
genres = ['religion','hobbies','lore','learned','romance']
modals = ['something','maybe','tomorrow','right','forever','to','from']

# Showing the distribution
cond_fdist.tabulate(conditions=genres, samples=modals)
```

```python
plt.figure(figsize=(14,6))
cond_fdist.plot()
```
